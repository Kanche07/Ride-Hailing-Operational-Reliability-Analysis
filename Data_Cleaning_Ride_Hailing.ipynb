```python
def null_values(df, name):
  nulls = df.isnull().sum()
  nulls = nulls[nulls>0]
  print(f"Null values in {name}:")
  print(nulls)

  print("-"*80)

null_values(drivers_df, "drivers_df")
null_values(sessions_df, "sessions_df")
null_values(events_df, "events_df")
null_values(trips_df, "trips_df")
null_values(disputes_df, "disputes_df")
```

```python
def duplicate_values(df, name):
    duplicates = df.duplicated().sum()
    print(f"Duplicate rows in {name}: {duplicates}")
    print("-"*80)


duplicate_values(drivers_df, "drivers")
duplicate_values(sessions_df, "app_sessions")
duplicate_values(events_df, "app_events")
duplicate_values(trips_df, "trips")
duplicate_values(disputes_df, "fare_disputes")
```
